```python
import matplotlib.pyplot as plt
import numpy as np

# 1. ë°ì´í„° ì„¤ì • (ì •í™•í•œ ê°’ ë°˜ì˜)
categories = ['Validation Accuracy (%)', 'Inference Latency (ms)', 'Model Size (MB)']
yamnet_values = [84.21, 37.17, 3.05]
panns_values = [20.05, 215.12, 311.22] # PANNs ì •í™•ë„ 20%ë¡œ ìˆ˜ì •ë¨

x = np.arange(len(categories))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 7))

# 2. ë§‰ëŒ€ ê·¸ë¦¬ê¸° (ë¡œê·¸ ìŠ¤ì¼€ì¼ ì ìš©)
rects1 = ax.bar(x - width/2, yamnet_values, width, label='YAMNet (Ours)', color='#4169E1') # ë¡œì–„ ë¸”ë£¨
rects2 = ax.bar(x + width/2, panns_values, width, label='PANNs (Baseline)', color='#B0C4DE') # ë¼ì´íŠ¸ ìŠ¤í‹¸ ë¸”ë£¨

# 3. ë¡œê·¸ ìŠ¤ì¼€ì¼ ì„¤ì • (ì¤‘ìš”!)
ax.set_yscale('log')
ax.set_ylim(1, 1000) # Yì¶• ë²”ìœ„ ì„¤ì •

# 4. ì¶• ë° ì œëª© ì„¤ì •
ax.set_xticks(x)
ax.set_xticklabels(categories, fontsize=12, fontweight='bold')
ax.legend(fontsize=12)
ax.set_title('YAMNet vs PANNs Performance Comparison', fontsize=16, fontweight='bold', pad=20)

# 5. ê°’ í‘œì‹œ í•¨ìˆ˜ (ë§‰ëŒ€ ìœ„ì— ìˆ«ì ì“°ê¸°)
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 5),
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=11, fontweight='bold')

autolabel(rects1)
autolabel(rects2)

# 6. ğŸ”¥ í•µì‹¬ ê°•ì¡° ë©˜íŠ¸ (Annotation) ì¶”ê°€
# (ì¢Œí‘œëŠ” ëˆˆëŒ€ì¤‘ìœ¼ë¡œ ë§ì¶¤, ë¡œê·¸ ìŠ¤ì¼€ì¼ì´ë¼ ìœ„ì¹˜ê°€ ì¢€ ë‹¤ë¥¼ ìˆ˜ ìˆìŒ)
# Accuracy
ax.annotate('4x Higher Accuracy', xy=(x[0]-width/2, 84.21), xytext=(x[0]-width/2, 150),
            arrowprops=dict(facecolor='red', shrink=0.05),
            ha='center', color='red', fontsize=11, fontweight='bold')

# Latency
ax.annotate('5.8x Faster', xy=(x[1]-width/2, 37.17), xytext=(x[1]-width/2, 8),
            arrowprops=dict(facecolor='green', shrink=0.05),
            ha='center', color='green', fontsize=11, fontweight='bold')

# Size
ax.annotate('100x Lighter', xy=(x[2]-width/2, 3.05), xytext=(x[2]-width/2, 1.5),
            arrowprops=dict(facecolor='green', shrink=0.05),
            ha='center', color='green', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.savefig('final_comparison_chart.png', dpi=300) # íŒŒì¼ë¡œ ì €ì¥
print("ì´ë¯¸ì§€ ì €ì¥ ì™„ë£Œ!")
plt.show()
```
